```python
IVET1 = pd.read_excel(r'C:\Users\lijunjie3\OneDrive\工作电脑文件\IVET 数据处理\4MEB_UAES_Base_plus_conti\IVET1\Conti._8000rpm_40Nm.xlsx')
```

```python
Std = Std.dropna()
IVET1 = IVET1.dropna()
datas = [Std, IVET1]
```

# 2. 转矩图对比

```python
linestyles = ['-', '-']
labels = ['Std', 'IVET1']
y = list()
for index, data in enumerate(datas):
    y.append({'data': [data['ManZeitIst'], data['M_HMmess']], 'style': {'linestyle': linestyles[index], 'label':labels[index]}})
```

```python
paras = {'x_label': "tims [s]",
             'y_label': "Torque [Nm]",
             'title': 'Continuous performance test (8000rpm, 40Nm, 65°C, 8L/min)'}
```

```python
fig_toruqe = base_plot(y, paras)
ax = fig_toruqe.axes[0]
ax.set_xlim((100, 3000));
ax.set_ylim((40, 43));
```

# 3. 三相平均电流对比

```python
linestyles = ['-', '-']
labels = ['Std', 'IVET1']
y = list()
for index, data in enumerate(datas):
    data['average_current'] = data[['PA1_IRMS_1', 'PA1_IRMS_2', 'PA1_IRMS_3']].mean(axis=1)
    y.append({'data': [data['ManZeitIst'], data['average_current']], 'style': {'linestyle': linestyles[index], 'label':labels[index]}})
```

```python
paras = {'x_label': "tims [s]",
             'y_label': "three phase current RMS average [A]",
             'title': 'Continuous performance test (8000rpm, 40Nm, 65°C, 8L/min)'}
```

```python
fig_toruqe = base_plot(y, paras)
ax = fig_toruqe.axes[0]
ax.set_xlim((100, 3000));
ax.set_ylim((80,100));
```
